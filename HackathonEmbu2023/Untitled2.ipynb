```python
import seaborn.objects as so
import seaborn as sns
```

```python
penguins=sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
(
  so.Plot(
      data = penguins,
      x = 'bill_length_mm',
      y = 'bill_depth_mm',
      color='species'
      
      
      )
    .add(
      so.Dot()
        
     
    )
    .add(
    so.Line(),
    so.PolyFit(order=1)
    )
    .facet(col='sex')
        
)
```

```python

(
    so.Plot(
        data=penguins,
        x="flipper_length_mm"
    )
    .add(
        so.Bar(),
        so.Hist()
    )
    .facet(row='species')
)
```

```python
(
  so.Plot(
      data = penguins,
      x = 'bill_depth_mm',
      y = 'species'
  )
  .add(
    so.Dots(),
    so.Jitter()
  )
  # Show the median
  .add(
    so.Dash(color='green'),
    so.Agg('median')
  )
    .add(
    so.Dash(color='red'),
    so.Agg('mean')
  )
  .layout(size=(5, 6))
    
)
```

```python
diamonds=sns.load_dataset("diamonds")
```

```python
(
  so.Plot(
    data = diamonds,
    x = 'carat',
    y = 'price'
  )
  .add(
    so.Dot(alpha = .05)
  )
  .layout(size=(5, 6))
)
```
